```python
# word tokensisation, creating each word as an element.
from nltk import word_tokenize
words = word_tokenize(text)
import string
len(words)
```

```python
string.punctuation
```

```python
# Removing punctuation symbols
punc = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
table = str.maketrans("","",punc)
```

```python
for w in words:
    stripped = [w.translate(table)]
```

```python
# OR Performing sentence,word tokenisation and punctuation removal in one step using regex library
import re
words = re.split(r'\W+',text)
```

```python
#lower caseing all words
words =[word.lower() for word in words ]
```

```python
# removing stopwords like is the an has their etc.
from nltk.corpus import stopwords
stop_words=  stopwords.words('english')
clean_words = [ w for w in words if w not in stop_words]
```

# Word cloud generation
Word clouds or text clouds or tag clouds are a very useful visual tool. It helps you have a cursory glance at your text data. The more a specific word appears in a text, the bigger and bolder it appears in the word cloud.

```python
# making a word cloud
def wordcloud(data, title):
    cloud = WordCloud(width=400,
                      height=330,
                      max_words=150,
                      colormap='tab20c',
                      collocations= True).generate(data)
    plt.figure(figsize=(10,8))
    plt.imshow(cloud)
    plt.axis('off')
    plt.title(title, fontsize=13)
    plt.show()
    
#Use the function to generate the word cloud
wordcloud(a, 'Doge to the moon')
```
