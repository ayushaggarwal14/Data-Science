```python

def wordcloud(data, title):
    cloud = WordCloud(width=400,
                      height=330,
                      max_words=150,
                      colormap='tab20c',
                      collocations= True).generate(data)
    plt.figure(figsize=(10,8))
    plt.imshow(cloud)
    plt.axis('off')
    plt.title(title, fontsize=13)
    plt.show()
    
#Use the function to generate the word cloud
wordcloud(a, 'Doge to the moon')
```

```python
import numpy as np
!pip install Pillow
import PIL
from PIL import Image
from numpy import asarray
image = Image.open("C:\\Users\\DELL\\Desktop\\AyushManu\\DataVader\\Assignments\\DogeCoin\\doge.jpg")
doge_mask = asarray(image)
```

```python
def doge_wordcloud(data, title, mask=None):
    cloud = WordCloud(scale=3,
                      colormap='RdYlGn',
                      mask=mask,
                      background_color='white',
                      collocations=True,
                      contour_color='black',
                      contour_width=1).generate(data)
    plt.figure(figsize=(14,8))
    plt.imshow(cloud)
    plt.axis('off')
    plt.title(title)
    plt.show()
    
# Use the function with the rome_corpus and our mask to create word cloud     
doge_wordcloud(a, 'Doge to the Moon', mask=doge_mask)
```
